```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Escalado de la imagen

def load_img():
    img=cv2.imread("Images/bricks.jpg").astype(np.float32)/255
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img

#Siempre estandarizar la informacion interna de la imagen.
#Normalizacion y escalado de la imagen.
```

```python
def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img)
```

```python
i=load_img()
display_img(i)

#El color que manda se puede saber con la generacion del histograma
```

```python
#Aclarar la imagen
img=load_img()
#gamma=1/4 #Propiedad para el brillo
gamma=2
effected_image=np.power(img,gamma)
display_img(effected_image)
```

```python
#Convolucion
#Aplicar una operacion de ajuste para aplicar un filtro en la imagen

img=load_img()
font=cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,text="OpenCV",org=(10,600),fontFace=font,fontScale=8,color=(255,0,0), thickness=3)
display_img(img)
```

```python
kernel=np.ones(shape=(5,5),dtype=np.float32)/25
```
