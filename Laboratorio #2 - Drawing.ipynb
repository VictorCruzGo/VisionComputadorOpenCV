```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Crear una matriz de 512x512 con canales RGB. Trabajar con valores enternos (dtype=np.int16)
black_img=np.zeros(shape=(512,512,3),dtype=np.int16)
black_img.shape
```

```python

plt.imshow(black_img)
```

```python
#Crear un rectangulo
cv2.rectangle(black_img,pt1=(384,0),pt2=(510,128),color=(0,255,0), thickness=5)
plt.imshow(black_img)

#Para crear un rectangulo se necesita dos puntos.
```

```python
cv2.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,255,0), thickness=5)
plt.imshow(black_img)
```

```python
#Crear una linea
cv2.line(black_img,pt1=(0,0),pt2=(511,511),color=(102,240,255),thickness=5)
plt.imshow(black_img)
```

```python
#Crear un table negro
canvas=np.zeros((300,300,3), dtype="uint8")
plt.imshow(canvas)
#El uint8, int16, es necesario para no consumir mucha memoria.
```

```python
#Crear un circulo
green=(0,255,0)
cv2.circle(canvas,(100,100),30,green,thickness=5)
plt.imshow(canvas)
```

```python
#Crear un table negro
canvas=np.zeros((300,300,3), dtype="uint8")
plt.imshow(canvas)
```

```python
canvas.shape
```

```python
#Crear circulos concentricos

white=(255,255,255)
(centerX,centerY)=(canvas.shape[1]//2, canvas.shape[0]//2)
for r in range(0,160,10):
    cv2.circle(canvas,(centerX,centerY),r,white,thickness=3)
plt.imshow(canvas)

#Los circulos concentricos se utiliza para: difuminar, ocultar informacion, ofuscacion
```

```python
#Crear un table negro
canvas=np.zeros((300,300,3), dtype="uint8")
plt.imshow(canvas)
```

```python
for i in range(0,25):
    radius = np.random.randint(5,high=200)
    color = np.random.randint(0,high=256,size=(3,)).tolist() #size=(3,), porque hay que tomar 3 colores
    pt=np.random.randint(0,high=300,size=(2,)) #size=(2,) poruqe un punto tiene dos coordenandas
    cv2.circle(canvas,tuple(pt),radius,color,-1)
plt.imshow(canvas)
```
