```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("Images/cat.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
num_rows, num_cols=img.shape[:2]
#rotation_matrix=cv2.getRotationMatrix2D((punto de rotacion),angulo,factor de escala).
rotation_matrix=cv2.getRotationMatrix2D((num_cols/2, num_rows/2),30,1)
img_rotation=cv2.warpAffine(img, rotation_matrix,(num_cols,num_rows)) #warpAffine, lleva acabo la translacion.
plt.imshow(img_rotation)

#Las rotaciones son necesarias para tener variedad de imagenes en el entrenamiento de la imagen.
```
