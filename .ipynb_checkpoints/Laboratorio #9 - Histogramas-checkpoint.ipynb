```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Los histogramas proporcionan los niveles de colores de una imagen.
```

```python
dark_horse=cv2.imread("Images/horse.jpg")
show_horse=cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB)
plt.imshow(show_horse)
```

```python
rainbow=cv2.imread("Images/rainbow.jpg")
show_rainbow=cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB)
plt.imshow(show_rainbow)
```

```python
blue_bricks=cv2.imread("Images/bricks.jpg")
show_blue_bricks=cv2.cvtColor(blue_bricks,cv2.COLOR_BGR2RGB)
plt.imshow(show_blue_bricks)
```

```python
##hist_values=cv2.calcHist([blue_bricks],'Canal azul'channels[0],'Normlamente el histograma se ve de toda la imagenen=None' mask=None,'Rango de la imagenes'histSize=[256],rango=[0,256])
hist_values=cv2.calcHist([blue_bricks],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
hist_values.shape
```

```python
#Histograma de la intensidad de azul.
plt.plot(hist_values)
```

```python
hist_values=cv2.calcHist([dark_horse],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
#Intensidades del color azul
#En la grafica no se aprecia ninguna intesidad de azul
plt.plot(hist_values)
```

```python
img=blue_bricks
color=('b','g','r')

for i,col in enumerate(color):
    hist=cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color=col)
    plt.xlim=([0,256]) #limites en el eje x
plt.title("Blue Bricks")
plt.show()
```

```python
img=dark_horse
color=('b','g','r')

for i,col in enumerate(color):
    hist=cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color=col)
    plt.xlim=([0,256]) #limites en el eje x
    plt.ylim=([0,100000])
plt.title("dark_horse")
plt.show()
```
