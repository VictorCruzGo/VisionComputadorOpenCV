```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread("Images/sammy_face.jpg")
```

```python
plt.imshow(img)
```

## Algoritmo Canny

```python
#Generalmente se utiliza el algoritmo de Cani para obtener bordes.
```

```python
#Obtener los bordes de la imagen
edges=cv2.Canny(image=img, threshold1=127, threshold2=127)

#Umbrales
#threshold1=127, threshold2=127
#Por recomendacion se va tener umbrales del punto central del espectro de colores de una imagen (0 a 255)
```

```python
plt.imshow(edges)
```

```python
#Deteccion de bordes con todo el espector de 0 a 255
edges=cv2.Canny(image=img, threshold1=0, threshold2=255)
```

```python
plt.imshow(edges)
```

```python
#Escoger correctamente el umbral (threshold)
#Calcular el valor medio del pixel
med_val=np.median(img)
med_val
```

```python
#Obtener el umbral menor y mayor utilizando recomendaciones matematicas.
lower = int(max(0,0.7*med_val)) #70% del valor medio obtenido. Valor mas bajo=0.
upper=int(min(255,1.3*med_val)) #1.3=30% del valor medio obtenido .Valor tope=255.
```

```python
edges=cv2.Canny(image=img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
lower
```

```python
upper
```

## Desenfocar la imagen para optimizar los bordes

```python
#Hay ocaciones cuando no se obtienen los bordes correctamente, 
#Es recomendable, primero desenfocar y despues trabajar con los bordes fuertes.
#Al desenfocar, el algoritmo de Canny solo se que con los bordes fuertes.
#Si Canny detecta muchos bordes, entonces es mejor desenforcar la imagen.
```

```python
blurred_img=cv2.blur(img,ksize=(5,5))
```

```python
edges=cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
#Incrementado el limite superior del umbral
edges=cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper+50)
plt.imshow(edges)
```
