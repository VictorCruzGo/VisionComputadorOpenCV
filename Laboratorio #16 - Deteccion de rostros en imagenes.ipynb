## Deteccion del rostro

```python
#Es un algoritmo de clasificacion
face_cascade=cv2.CascadeClassifier("haarcascades/haarcascade_frontalface_default.xml")
```

```python
def detect_face(img):
    face_img=img.copy()
    face_rects=face_cascade.detectMultiScale(face_img,scaleFactor=1.2, minNeighbors=5)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    return face_img
```

```python
result=detect_face(nadia)
```

```python
plt.imshow(result,cmap="gray")
```

## Deteccion de los ojos

```python
eye_cascade=cv2.CascadeClassifier("haarcascades/haarcascade_eye.xml")
```

```python
def detect_eye(img):
    face_img=img.copy()
    #Hace el barrido en la imagen para detectar los ojos
    #eye_rects=eye_cascade.detectMultiScale(face_img) 
    #scaleFactor va entre 1.2 y 1.3
    #minNeighbors, numero impart. Por recomendacion el valor 5 
    eye_rects=eye_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5) 
    for (x,y,w,h) in eye_rects:
        cv2.rectangle(face_img, (x,y),(x+w,y+h),(255,255,255),10)
    return face_img
```

```python
result=detect_eye(nadia)
```

```python
plt.imshow(result,cmap="gray")
```
