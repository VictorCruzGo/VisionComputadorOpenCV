```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img =cv2.imread("Images/input.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#Trasladar la imagen con recorte
num_rows, num_cols = img.shape[:2]
#translation_matrix = np.float32([[1,0,Tx],[0,1,Ty]])
translation_matrix = np.float32([[1,0,150],[0,1,110]])
img_translation = cv2.warpAffine(img,translation_matrix,(num_cols,num_rows),cv2.INTER_LINEAR)
plt.imshow(img_translation)
```

```python
#Transladar la imagen completa
img =cv2.imread("Images/input.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
num_rows, num_cols = img.shape[:2]
translation_matrix=np.float32([[1,0,70],[0,1,110]])
img_translation=cv2.warpAffine(img_translation,translation_matrix,(num_cols+70,num_rows+110))

translation_matrix=np.float32([[1,0,-30],[0,1,-50]])
img_translation=cv2.warpAffine(img_translation,translation_matrix,(num_cols+70+30,num_rows+110+50))

plt.imshow(img_translation)

#Extrapolacion de pixeles
#Llenar de graficos en horizontal y vertical
```
