```python
import cv2 #Libreria de openCV.
import numpy as np #Libreria para calculos numericos.
import matplotlib.pyplot as plt #Libreria para graficos.
%matplotlib inline
```

```python
#Cargar imagen
img = cv2.imread("Images/input.jpg")
#Mostrar imagen
plt.imshow(img)

#OpenCV cambia el orden de los colores. Trabaja en BGR
```

```python
#Modificando el espacio de colores de BGR a RGB
img_fix= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_fix)
```

```python
#Cargar imagen en escala de grises
img_gray=cv2.imread("Images/input.jpg", cv2.IMREAD_GRAYSCALE)
#plt.imshow(img_gray,cmap="gray")
plt.imshow(img_gray, cmap="gray")

#El procesamiento de imagenes siempre se lo trabaja en escala de grises.
#Es mas rapido y usa menos recursos el procesamiento en escala de grises.
```

```python
#Guardar una imagen de grises en un nuevo archivo
cv2.imwrite("Images/input_gray.jpg",img_gray)
```

```python
cv2.imwrite("Images/input_gray.png",img_gray)
```

```python
cv2.imwrite("Images/input_gray_compression.png",img_gray,[cv2.IMWRITE_PNG_COMPRESSION])
```

## Manipulacion de pixeles y areas

```python
#Obtener los pesos de colores de la posicion[0,0]
#Pesos de colores RGB en la posicion [0,0] de la imagen
(b,g,r)=img[0,0]
print("pixeles en canal Rojo:{} - Verde:{} - Azul:{}".format(r,g,b))
```

```python
#Asigar a la posicon[0,0] de la imagen
img[0,0]=(0,0,255)
(b,g,r)=img[0,0]
print("pixeles en canal Rojo:{} - Verde:{} - Azul:{}".format(r,g,b))
```
